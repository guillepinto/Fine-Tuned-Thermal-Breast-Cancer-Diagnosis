```python
import numpy as np

# Lista de diccionarios con todos los modelos
models = [
    vit_32_full_00001, 
    resnet_16_full_0001, 
    resnet_16_full_00001, 
    vit_64_classifierconv1_00001, 
    xception_64_classifierconv1_0001, 
    resnet_32_full_00001_sgd
]

# Diccionario para almacenar los resultados
all_results = {}

# Función para calcular los promedios y desviación estándar
def calculate_metrics(model):
    metrics = ['test_loss', 'test_accuracy', 'test_recall', 'test_precision', 'test_f1', 'train_loss', 'train_accuracy']
    results = {}

    for metric in metrics:
        values = []

        # Iterar sobre los folds para extraer los valores
        for fold in range(1, 8):  # Cambia si tienes más o menos folds
            fold_key = f"fold_{fold}"
            if fold_key in model:
                values.append(model[fold_key][metric])

        # Calcular promedio y desviación estándar
        mean_value = np.mean(values)
        std_value = np.std(values)

        results[metric] = {
            'mean': mean_value,
            'std': std_value
        }

    return results

# Iterar sobre todos los modelos
for model in models:
    model_name = f"{model['architecture']}_{model['batch_size']}_{model['fine_tune']}_{model['learning_rate']}_{model['optimizer']}"
    all_results[model_name] = calculate_metrics(model)

# Imprimir los resultados para todos los modelos
for model_name, metrics in all_results.items():
    print(f"Resultados para {model_name}:")
    for metric, stats in metrics.items():
        print(f"  {metric}: {stats['mean']:.4f} (+/- {stats['std']:.4f})")
    print()  # Espacio entre modelos
```
